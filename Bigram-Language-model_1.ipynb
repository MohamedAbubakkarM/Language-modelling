```python
'''
  We're creating a Bigram language model, that works by predicting the next character from preceding character.
  That is we're working by character level.
'''
b = {}

for w in words:
  # For language model to recognize the start and end of a word, we intentionally include them for every words
  # So that the LM can understand that this character can be used at start of a name and end of a name.
  # Store every two characters in a dict to have the frequency of their occurence.
  chs = ['<S>'] + list(w) + ['<E>']
  for ch1, ch2 in zip(chs, chs[1:]):
    print(ch1, ch2)
    bigram = (ch1, ch2)
    b[bigram] = b.get(bigram, 0) + 1
```

```python
# Creating a lookup table to map in a Tensor
chars = sorted(list(set(''.join(words).lower())))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
itos
```

```python
import torch
N = torch.zeros((27,27), dtype=torch.int32)
```

```python
for w in words:
  chs = ['.'] + list(w.lower()) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    N[ix1, ix2] += 1
```

```python
#We are going to do sampling from the model. For that, we need to normalise the data values
#These are probability distribution, which tells us the probability of a character appearing start of a word (We are considering N[0]--which has starting characters.)
p = N[0].float()
p /= p.sum()
p.sum()

#Sampling from the first row of the entire tensor
g = torch.Generator().manual_seed(2147483647)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
itos[ix]
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(20):
    out = []

    ix = 0
    while True:
      p = N[ix].float()
      p /= p.sum()
      ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
      out.append(itos[ix])
      if ix == 0:
        break
    print(''.join(out))
```

```python
# Instead of doing manually like selecting a row and normalizing it and then sampling from it, we make the entire tensor normalized
P = N.float()
# Here, if we do P.sum(), it gives us the sum of the entire tensor, but we want to normalize every rows based on the current row's sum
# For that, we have to mention the axis of doing the sum
P.sum(0, keepdim=True).shape # This tensor contains sum of every columns
# The shape is [1, 27] because, the entire 27 columns are mapped to a single row with 27 columns, becomes a row vector.
# doing with 1 makes it good for row wise sum, now it becomes a column vector
P /= P.sum(1, keepdim=True)
P[0]
```

```python
W = torch.randn((27, 27), generator=g)
```

```python
xs, ys = [], []

log_likelihood = 0.0

for w in words[:1]:
  chs = ['.'] + list(w.lower()) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    print(ch1, ch2)
    prob = P[ix1, ix2]
    log_prob = torch.log(prob)
    print(f"{prob}: {log_prob}")
    log_likelihood += log_prob
    xs.append(ix1)
    ys.append(ix2)
print(f"{log_likelihood=}")

xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
import torch.nn.functional as F
xenc = F.one_hot(xs, num_classes=27).float()
W = torch.randn((27, 27), generator=g, requires_grad=True) # Each neuron receives 27 inputs.
# We can't just simply input the count for NN. So, similar to count, we are assign log count, which is e^x
# We call it as logits
```

```python
logits = xenc @ W # Input to the network
counts = logits.exp() # Equivalent to N
probs = counts / counts.sum(1, keepdims=True) # Probabilities for next character
#The above two lines are equivalent to applying a softmax function
probs # These are the probabilities coming out of neural net
```

```python
loss = -probs[torch.arange(5), ys].log().mean()
loss.item()
# We have to reduce this loss. So, we decided to backpropagate and adjust the gradient
```

```python
W.grad = None
loss.backward()
```

```python
W.data += -0.1 * W.grad
```

```python
xs, ys = [], []

for w in words:
  chs = ['.'] + list(w.lower()) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    xs.append(ix1)
    ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
num
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, requires_grad=True)
```

```python
# Gradient Descent

for k in range(10):

  xenc = F.one_hot(xs, num_classes=27).float()
  logits = xenc @ W
  counts = logits.exp()
  probs = counts / counts.sum(1, keepdims=True)
  loss = -probs[torch.arange(num), ys].log().mean() + 0.01*(W**2).mean() # Regularization happens here, same as model smoothing
  print(loss.item())

  # backward pass
  W.grad = None
  loss.backward()

  W.data += -50 * W.grad
```

```python
for i in range(5):
  out = []
  ix = 0
  while True:
    # p = P[ix]
    xenc = F.one_hot(torch.tensor([ix]), num_classes=27).float()
    logits = xenc @ W
    counts = logits.exp()
    p = counts / counts.sum(1, keepdims=True)
    ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    if ix == 26: continue
    out.append(itos[ix])
    if ix == 0:
      break
  print(''.join(out))
```
